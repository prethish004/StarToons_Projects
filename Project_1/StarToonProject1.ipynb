**Peak Detection without inbuild function**



```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.loadtxt('/content/Data_1.txt')
data2 = np.loadtxt('/content/Data_2.txt')
```

```python
def find_max_min(data):
    max = []
    min = []

    for i in range(1, len(data) - 1):
        if data[i-1] < data[i] > data[i+1]:
            max.append(i)
        if data[i-1] > data[i] < data[i+1]:
            min.append(i)
    return max, min
peaks1, valleys1 = find_max_min(data1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data1, label='Data-1')
plt.plot(peaks1, data1[peaks1], 'ro')
plt.plot(valleys1, data1[valleys1], 'bo')
plt.title('Data-1')
plt.show()
```

```python
print("Max for Data_1: ", peaks1)
print("Min for Data_1: ", valleys1)
```

```python
peaks2, valleys2 = find_max_min(data2)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data2, label='Data-2')
plt.plot(peaks2, data2[peaks2], 'ro')
plt.plot(valleys2, data2[valleys2], 'bo')
plt.title('Data-2')
plt.show()
```

```python
print("Max for Data_2: ", peaks2)
print("Min for Data_2: ", valleys2)
```

**Peak Detection with inbuild function**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

data_1 = np.loadtxt('/content/Data_1.txt')
data_2 = np.loadtxt('/content/Data_2.txt')
```

```python
peaks1= find_peaks(data_1)[0]
valleys1 = find_peaks(-data_1)[0]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data_1, label='Data_1')
plt.plot(peaks1, data_1[peaks1], 'ro')
plt.plot(valleys1, data_1[valleys1], 'bo')
plt.title('Data_1')
plt.show()
```

```python
print("Max for Data_1: ", peaks1)
print("Min for Data_1: ", valleys1)
```

```python
peaks2= find_peaks(data_2)[0]
valleys2 = find_peaks(-data_2)[0]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data_2, label='Data_2')
plt.plot(peaks2, data_2[peaks2], 'ro')
plt.plot(valleys2, data_2[valleys2], 'bo')
plt.title('Data_2')
plt.show()
```

```python
print("Max for Data_2: ", peaks2)
print("Min for Data_2: ", valleys2)
```

**By adding Minimal Distance value**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

data_1 = np.loadtxt('/content/Data_1.txt')
data_2 = np.loadtxt('/content/Data_2.txt')
```

```python
min_distance = 40
peaks1 = find_peaks(data_1, distance=min_distance)[0]
valleys1 = find_peaks(-data_1, distance=min_distance)[0]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data_1, label='Data_1')
plt.plot(peaks1, data_1[peaks1], 'ro', label='Max')
plt.plot(valleys1, data_1[valleys1], 'bo', label='Min')
plt.title('Data_1')
plt.legend()
plt.show()
```

```python
print("Max for Data_1: ", peaks1)
print("Min for Data_1: ", valleys1)
```

```python
peaks2= find_peaks(data_2, distance=min_distance)[0]
valleys2 = find_peaks(-data_2, distance=min_distance)[0]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data_2, label='Data-2')
plt.plot(peaks2, data_2[peaks2], 'ro')
plt.plot(valleys2, data_2[valleys2], 'bo')
plt.title('Data-2')
plt.show()
```

```python
print("Max for Data_2: ", peaks2)
print("Min for Data_2: ", valleys2)
```
